```python
# use this for sankeys with >2 levels
df6 = ovi_HB_chunk[['0.75','1.0','counts']].groupby(['0.75','1.0']).count().reset_index()
df6.columns = ['source','target','value']
df6
```

```python
# this is basically a mapping dictionary of nodes enumerated
mapping_dict = {k: v for v, k in enumerate(nodes)}
mapping_dict
```

```python
# replace source and target with enumerated values
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)
links
```

```python
# turn this table into a dictionary for making the sankey diagram
links_dict = links.to_dict(orient='list')
links_dict
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        line=dict(color='blue', width=0.5),
        label = nodes,
        color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])
fig.update_layout(title='Prominent Inputs to oviINr in the full ovi modularity data across resolutions')
fig.show()
```

```python
# append _r0.0 to values in column '0.0' and so on
ovi_out_chunk['0.0'] = ovi_out_chunk['0.0'].astype(str) + '_r0.0r_out'
ovi_comboout_chunk['0.0'] = ovi_comboout_chunk['0.0'].astype(str) + '_r0.0combo_out'
```

Here, I have to merge the dfs. Doing an inner merge will basically shed any bodyID that is not in both dataframes. 

```python
# merge the dataframes on the id column
ovi_merged_chunk = pd.merge(ovi_out_chunk, ovi_comboout_chunk, on='id', how='inner')
ovi_merged_chunk
```

```python
# these are the nodes that will be used in the sankey diagram
nodes = ovi_merged_chunk['0.0_x'].unique().tolist() + ovi_merged_chunk['0.0_y'].unique().tolist()
```

```python
# add a column of ones to ovi_HB_in_chunk
ovi_merged_chunk['counts'] = 1
```

```python
# use this for sankeys with only 2 levels
df2 = ovi_merged_chunk.groupby(['0.0_x','0.0_y'])['counts'].count().reset_index()
df2.columns = ['source','target','value']
```

```python
links = df2
```

```python
# this is basically a mapping dictionary of nodes enumerated
mapping_dict = {k: v for v, k in enumerate(nodes)}
```

```python
# replace source and target with enumerated values
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)
```

```python
# turn this table into a dictionary for making the sankey diagram
links_dict = links.to_dict(orient='list')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        #line=dict(color='blue', width=0.5),
        label = nodes,
        #color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])
fig.update_layout(title='mapping from oviINr output to combined oviIN output connectome coarse modularity', height=1000)
fig.show()
```
