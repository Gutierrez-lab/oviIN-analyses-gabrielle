```python
ovi_HB_chunk[['0.0','0.05','counts']].groupby(['0.0','0.05']).count().reset_index()
```

```python
# use this for sankeys with only 2 levels
df2 = ovi_HB_chunk.groupby(['0.0','0.05'])['Customer'].count().reset_index()
df2.columns = ['source','target','value']
df2
```

```python
# use this for sankeys with >2 levels
df2 = ovi_HB_chunk[['0.0','0.05','counts']].groupby(['0.0','0.05']).count().reset_index()
df2.columns = ['source','target','value']
df2
```

```python
# use this for sankeys with >2 levels
df3 = ovi_HB_chunk[['0.05','0.5','counts']].groupby(['0.05','0.5']).count().reset_index()
df3.columns = ['source','target','value']
df3
```

```python
links = pd.concat([df2, df3], axis=0)
links
```

```python
# this is basically a mapping dictionary of nodes enumerated
mapping_dict = {k: v for v, k in enumerate(nodes)}
mapping_dict
```

```python
links = df2.copy()
# below is not needed since I have nodes
unique_source_target = list(pd.unique(links[['source','target']].values.ravel('k')))
unique_source_target
```

```python
# replace source and target with enumerated values
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)
links
```

```python
# turn this table into a dictionary for making the sankey diagram
links_dict = links.to_dict(orient='list')
links_dict
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        line=dict(color='blue', width=0.5),
        label = nodes,
        color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])
fig.update_layout(title='Prominent Inputs to oviINr in the full ovi modularity data across resolutions')
fig.show()
```

```python
nodes = ["PhD", "Career Outside Science",  "Early Career Researcher", "Research Staff",
         "Permanent Research Staff",  "Professor",  "Non-Academic Research"]
nodes = hv.Dataset(enumerate(nodes), 'index', 'label')
edges = [
    (0, 1, 53), (0, 1, 25), (0, 2, 47), (2, 6, 17), (2, 3, 30), (3, 1, 22.5), (3, 4, 3.5), (3, 6, 4.), (4, 5, 0.45)   
]
```

```python

value_dim = hv.Dimension('Percentage', unit='%')
careers = hv.Sankey((edges, nodes), ['From', 'To'], vdims=value_dim)

careers.opts(
    opts.Sankey(labels='label', label_position='right', width=900, height=300, cmap='Set1',
                edge_color=dim('To').str(), node_color=dim('index').str()))
```
