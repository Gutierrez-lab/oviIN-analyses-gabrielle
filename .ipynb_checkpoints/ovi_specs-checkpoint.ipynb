## oviIN's personal specs

```python
# inputs to oviINr
from neuprint import fetch_simple_connections
ovi_inputs = fetch_simple_connections(None,oviINr_bodyID)
ovi_inputs
```

The number of unique bodyIDs pre-synaptic to oviINr.

```python
ovi_inputs['bodyId_pre'].nunique()
```

The number of total synaptic sites from pre-synaptic neurons onto oviINr. This number is slightly lower than the number given by the Neuprint GUI (23029).

```python
ovi_inputs['weight'].sum()
```

Interestingly and thankfully, fetch_adjacencies gives the same results when include_nonprimary=False. 

```python
from neuprint import fetch_adjacencies
n_df, conn_df = fetch_adjacencies(None,oviINr_bodyID,include_nonprimary=False)
```

```python
conn_df
```

```python
conn_df['bodyId_pre'].nunique()
```

```python
conn_df['weight'].sum()
```

```python
# outputs from oviINr
from neuprint import fetch_simple_connections
ovi_outputs = fetch_simple_connections(oviINr_bodyID,None)
ovi_outputs
```

The number of unique bodyIDs post-synaptic to oviINr

```python
ovi_outputs['bodyId_post'].nunique()
```

The number of total post-synaptic sites from oviINr. This number is much higher than that given in Neuprint (6863). Somehow, they must've filtered something out. Maybe I should disregard the GUI because nothing is really making sense there...

```python
ovi_outputs['weight'].sum()
```

```python
from neuprint import fetch_adjacencies
n_df, conn_df = fetch_adjacencies(oviINr_bodyID,None,include_nonprimary=False)
conn_df
```

```python
conn_df['bodyId_post'].nunique()
```

```python
conn_df['weight'].sum()
```

The total number of direct partners for oviINr.

```python
ovi_partners = pd.concat([ovi_inputs['bodyId_pre'],ovi_outputs['bodyId_post']],ignore_index=True)
ovi_partners.nunique()
```

## Where the synapses are
Getting descriptive about oviIN's synaptic sites. Not only how many, but where they are located.

The ROIs for synapses that oviIN receives.

```python
from neuprint import fetch_synapses, NeuronCriteria as NC, SynapseCriteria as SC

fetch_synapses(oviINr_bodyID, SC(type='post',primary_only=True))
```

Interestingly, these results match up with Neuprint's results for synapse counts whereas the simple_connections and fetch_adjacencies counts differ. Unclear why those are different. 

These are synapses received by oviIN retrieved by fetch_synapse_connections which provides information about the connecting neuron. It returns about 2000 fewer synapses. Perhaps some of the post-syn sites on oviIN don't have a partner and only partnered connections are retrieved here?

```python
from neuprint import fetch_synapse_connections

fetch_synapse_connections(None, oviINr_bodyID, SC(primary_only=True))
```

The ROIs for synapses that oviIN projects to other neurons.

```python
from neuprint import fetch_synapses, NeuronCriteria as NC, SynapseCriteria as SC

fetch_synapses(oviINr_bodyID, SC(type='pre',primary_only=True))
```

Getting ovi outputs using fetch_synapse_connections times out but it appears that it was going to return 29274 connections which is what fetch_simple_connections and fetch_adjacencies returns. 

```python
from neuprint import fetch_synapse_connections

fetch_synapse_connections(oviINr_bodyID, None, SC(primary_only=True))
```
