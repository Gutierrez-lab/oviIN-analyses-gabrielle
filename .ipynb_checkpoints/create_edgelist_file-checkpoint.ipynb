From prameshsingh's instructions: "Use an edgelist file including 3 columns (separated by space or tab with no header) with the last one representing the weight. (If the network is unweighted, just put 1 to the whole column. The network should be undirected. Self-loops will be ignored. See example in karate.txt)"

The table created below will be for inputs to oviIN. 

```python
# create a table with a column that is oviIN's bodyID, a column that has the bodyIDs of its inputs, and a column of weights

from neuprint import fetch_adjacencies

neuron_df, conn_df = fetch_adjacencies(None, oviINr_bodyID)
```

```python
edgelist = conn_df[['bodyId_pre','bodyId_post','weight']]
```

```python
# export this edgelist without indices or headers
edgelist.to_csv('ovi_edgelist.txt', sep=' ', header=False, index=False)
```
