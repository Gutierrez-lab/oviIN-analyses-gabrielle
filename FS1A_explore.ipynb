```python
from neuprint import fetch_simple_connections
# get inputs to FS1A
FS1A_inputs = fetch_simple_connections(None,'FS1A')
FS1A_inputs
```

```python
# group by the presynaptic type and sum up the weights of each type
FS1A_input_types = FS1A_inputs.groupby(['type_pre'], as_index=False)['weight'].sum()

# display results sorted by highest to lowest weight
FS1A_input_types = FS1A_input_types.sort_values(by='weight', ascending=False)
FS1A_input_types.head(20)
```

```python
from neuprint import fetch_simple_connections
# get inputs to FS1A
FS1A_outputs = fetch_simple_connections('FS1A',None)
FS1A_outputs
```

```python
# group by the postsynaptic type and sum up the weights of each type
FS1A_output_types = FS1A_outputs.groupby(['type_post'], as_index=False)['weight'].sum()

# display results sorted by highest to lowest weight
FS1A_output_types = FS1A_output_types.sort_values(by='weight', ascending=False)
FS1A_output_types.head(20)
```
