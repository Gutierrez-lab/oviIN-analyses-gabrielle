Total counts gets the counts for items in col1 and col2. In this example, that would be the counts of nodes in each cluster for 0.0 resolution and counts of cell types.

```python
total_counts = {c: df.value_counts(c) for c in [col1, col2]}
total_counts
```

Joint does a joint count of items in those columns such that, for example, the first row has a count of the number of FC2C cell types in cluster 3. 

```python
joint = df.value_counts([col1, col2]).reset_index().rename(columns={0:'count'})
joint
```

This next piece of code puts the total counts for each cluster into the joint df. The count column is $n_{C \cap T}$ and col1_count is $n_C$ which is the number of cells in a cluster.

```python
joint['col1_count'] = total_counts[col1][joint[col1]].values
joint
```

The joint counts are divided by the cluster counts to get the proportions of cell types within each cluster. Thus, $p = \frac{n_{C \cap T}}{n_C}$.

```python
p = joint['count'] / joint['col1_count']
p
```

The conditional entropy is $-\frac{n_{C \cap T}}{n_C} log2(\frac{n_{C \cap T}}{n_C})$ which is not technically conditional entropy which would be $-n_{C \cap T} log2(\frac{n_{C \cap T}}{n_C})$. 

```python
joint['conditional_entropy'] = -p * np.log2(p)
joint
```

res sums up the (almost) conditional entropies for each cluster and counts the number of cell types in each cluster.

```python
res = joint.groupby(col1).agg({'conditional_entropy': 'sum', col2:'count'})
res
```

A heterogeneity column is added to the res df. Wherever conditional entropy is 0, put that value into the heterogeneity column. Otherwise, divide conditional entropy by $log2(n_{T_C})$ and put that value into the heterogeneity column. $n_{T_C}$ is the number of cell types in a given cluster. I believe this operation normalizes the conditional entropy. 

```python
res['heterogeneity'] = res['conditional_entropy'].where(res['conditional_entropy'] == 0,res['conditional_entropy'] / np.log2(res[col2]))
res
```

The remaining lines rename the celltype column and puts $n_C$ into its own column.

```python
res = res.rename(columns={col2:'n_'+col2})
res[col1+'_count'] = total_counts[col1][res.index]
res
```

Here is Alex's function. 

```python
def heterogeneity_score(df, col1, col2):
    """Compute the heterogeneity of `col1` with respect to `col2`.
    The idea is, each column represents a different labeling of, e.g. nodes in a graph.
    This computes the conditional entropy of the distribution of `col2` given `col1`,
    and returns
    """
    total_counts = {c: df.value_counts(c) for c in [col1, col2]}
    
    joint = df.value_counts([col1, col2]).reset_index().rename(columns={0:'count'})
    joint['col1_count'] = total_counts[col1][joint[col1]].values
    p = joint['count'] / joint['col1_count']
    joint['conditional_entropy'] = -p * np.log2(p)
    res = joint.groupby(col1).agg({'conditional_entropy': 'sum', col2:'count'})
    res['heterogeneity'] = res['conditional_entropy'].where(res['conditional_entropy'] == 0,
                                                            res['conditional_entropy'] / np.log2(res[col2]))
#     res['heterogeneity'] = res['conditional_entropy'] / np.log2(res[col2])
#     res['heterogeneity'].fillna(0, inplace=True)
    res = res.rename(columns={col2:'n_'+col2})
    res[col1+'_count'] = total_counts[col1][res.index]
    return res
```

# Completeness
"Cluster completeness measures how well represented each cell type is in a given cluster. This is the fraction of a cell type present in a cluster averaged across the cell types within that cluster, defined by the following:"
$Com(C) = \sum_T \frac{n_{C \cap T}}{n_C} \frac{n_{C \cap T}}{n_T}$

We alread have the first term in the variable p. 

```python
p = joint['count'] / joint['col1_count']
p
```

col2_count ends up being nT, i.e. the number of cells of type T in the whole network.

```python
joint['col2_count'] = total_counts[col2][joint[col2]].values
joint
```

q is the second term: $\frac{n_{C \cap T}}{n_T}$

```python
q = joint['count'] / joint['col2_count']
q
```

```python
joint['completeness'] = p * q
joint
```

resc sums up the (almost) completeness for each cluster and counts the number of cell types in each cluster.

```python
resc = joint.groupby(col1).agg({'completeness': 'sum', col2:'count'})
resc
```
