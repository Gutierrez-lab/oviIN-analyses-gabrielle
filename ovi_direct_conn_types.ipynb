```python
# body IDs of oviINs from Neuprint
oviINr_bodyID = 423101189
oviINl_bodyID = 485934965
```

Doing the inputs first.

```python
# inputs to oviINr
from neuprint import fetch_simple_connections
ovi_inputs = fetch_simple_connections(None,[oviINr_bodyID,oviINl_bodyID])
ovi_inputs
```

Unfortunately, this sums the bodyIds. Ideally, I'd like to list the bodyIds within a column, json style.

```python
ovi_in_types = ovi_inputs.groupby('type_pre').sum('weight')
ovi_in_types.sort_values(by="weight", ascending=False)
```

```python
ovi_in_bId_lists = ovi_inputs.groupby('type_pre')['bodyId_pre'].apply(list).reset_index()#.sum('weight')
ovi_in_bId_lists
```

```python
ovi_in_summed_weights = ovi_inputs.groupby('type_pre').sum('weight')
ovi_in_summed_weights
```

```python
ovi_input_types = ovi_in_bId_lists.merge(ovi_in_summed_weights[['weight']], on='type_pre').sort_values(by="weight", ascending=False)
ovi_input_types
```

```python
ovi_input_types.to_csv('ovi_input_types.csv')
```

Now the outputs.

```python
# outputs to oviINr
from neuprint import fetch_simple_connections
ovi_outputs = fetch_simple_connections([oviINr_bodyID,oviINl_bodyID],None)
ovi_outputs
```

```python
ovi_out_bId_lists = ovi_outputs.groupby('type_post')['bodyId_post'].apply(list).reset_index()#.sum('weight')
ovi_out_bId_lists
```

```python
ovi_out_summed_weights = ovi_outputs.groupby('type_post').sum('weight')
ovi_out_summed_weights
```

```python
ovi_output_types = ovi_out_bId_lists.merge(ovi_out_summed_weights[['weight']], on='type_post').sort_values(by="weight", ascending=False)
ovi_output_types
```

```python
ovi_output_types.to_csv('ovi_output_types.csv')
```
