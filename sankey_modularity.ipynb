```python
# use this for sankeys with >2 levels
df5 = ovi_HB_chunk[['0.5','0.75','counts']].groupby(['0.5','0.75']).count().reset_index()
df5.columns = ['source','target','value']
df5
```

```python
# use this for sankeys with >2 levels
df6 = ovi_HB_chunk[['0.75','1.0','counts']].groupby(['0.75','1.0']).count().reset_index()
df6.columns = ['source','target','value']
df6
```

```python
# replace source and target with enumerated values
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)
links
```

```python
# turn this table into a dictionary for making the sankey diagram
links_dict = links.to_dict(orient='list')
links_dict
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        line=dict(color='blue', width=0.5),
        label = nodes,
        color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])
fig.update_layout(title='Prominent Inputs to oviINr in the full ovi modularity data across resolutions')
fig.show()
```

## Grand Sankey
This is a Sankey of the complete modularity data for oviINr's full connectome.

```python
# take only the columns with modularity data
ovi_HB_chunk = ovi_HB_node_df[['0.0','0.05','0.1','0.5','0.75','1.0']].copy()
```

```python
# append _r0.0 to values in column '0.0' and so on
ovi_HB_chunk['0.0'] = ovi_HB_chunk['0.0'].astype(str) + '_r0.0'
ovi_HB_chunk['0.05'] = ovi_HB_chunk['0.05'].astype(str) + '_r0.05'
ovi_HB_chunk['0.1'] = ovi_HB_chunk['0.1'].astype(str) + '_r0.1'
ovi_HB_chunk['0.5'] = ovi_HB_chunk['0.5'].astype(str) + '_r0.5'
ovi_HB_chunk['0.75'] = ovi_HB_chunk['0.75'].astype(str) + '_r0.75'
ovi_HB_chunk['1.0'] = ovi_HB_chunk['1.0'].astype(str) + '_r1.0'

# add a column of ones to ovi_HB_chunk
ovi_HB_chunk['counts'] = 1
ovi_HB_chunk
```

```python
# these are the nodes that will be used in the sankey diagram
nodes = ovi_HB_chunk['0.0'].unique().tolist() + ovi_HB_chunk['0.05'].unique().tolist() + ovi_HB_chunk['0.1'].unique().tolist() + ovi_HB_chunk['0.5'].unique().tolist() + ovi_HB_chunk['0.75'].unique().tolist() + ovi_HB_chunk['1.0'].unique().tolist()
```

```python
# use this for sankeys with only 2 levels
df2 = ovi_HB_chunk.groupby(['0.0','0.05'])['counts'].count().reset_index()
df2.columns = ['source','target','value']
df2
```

```python
# use this for sankeys with >2 levels
df3 = ovi_HB_chunk[['0.05','0.1','counts']].groupby(['0.05','0.1']).count().reset_index()
df3.columns = ['source','target','value']

df4 = ovi_HB_chunk[['0.1','0.5','counts']].groupby(['0.1','0.5']).count().reset_index()
df4.columns = ['source','target','value']

df5 = ovi_HB_chunk[['0.5','0.75','counts']].groupby(['0.5','0.75']).count().reset_index()
df5.columns = ['source','target','value']

df6 = ovi_HB_chunk[['0.75','1.0','counts']].groupby(['0.75','1.0']).count().reset_index()
df6.columns = ['source','target','value']
```

```python
links = pd.concat([df2, df3, df4, df5, df6], axis=0)
links
```

```python
# this is basically a mapping dictionary of nodes enumerated
mapping_dict = {k: v for v, k in enumerate(nodes)}
mapping_dict
```

```python
# replace source and target with enumerated values
links['source'] = links['source'].map(mapping_dict)
links['target'] = links['target'].map(mapping_dict)
links
```

```python
# turn this table into a dictionary for making the sankey diagram
links_dict = links.to_dict(orient='list')
links_dict
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        #line=dict(color='blue', width=0.5),
        label = nodes,
        #color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])
fig.update_layout(title='full oviINr modularity data across resolutions', height=1000)
fig.show()
```

Cluster 3 in res 0.0 seems to maintain its integrity through increasing resolutions with only smaller parts breaking off. A core part of most of the other 0.0 clusters remains through res 0.1. It is hard to imagine that the higher resolutions contain anything meaningful. However, the Sankey with just the prominent inputs indicates that those inputs stick together in their groups throughout the resolutions until 1.0. 

```python
# threshold the data to exclude singletons and doubletons
links = links[links['value'] > 10]

# turn this table into a dictionary for making the sankey diagram
links_dict = links.to_dict(orient='list')
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
        thickness=20,
        #line=dict(color='blue', width=0.5),
        label = nodes,
        #color='green'
    ),
    link = dict(
        source= links_dict['source'],
        target = links_dict['target'],
        value = links_dict['value']
    )
    
)
])
fig.update_layout(title='full oviINr modularity data across resolutions', height=1000)
fig.show()
```

With a thresholding of 10, the little pieces that break off are cleared away. Here you can see that very little remains beyond res 0.75. At res 0.1 there is a big break where things went off into singletons and into smaller groups. There are some interesting looking recombinations at the higher resolutions.

```python
nodes = ["PhD", "Career Outside Science",  "Early Career Researcher", "Research Staff",
         "Permanent Research Staff",  "Professor",  "Non-Academic Research"]
nodes = hv.Dataset(enumerate(nodes), 'index', 'label')
edges = [
    (0, 1, 53), (0, 1, 25), (0, 2, 47), (2, 6, 17), (2, 3, 30), (3, 1, 22.5), (3, 4, 3.5), (3, 6, 4.), (4, 5, 0.45)   
]
```

```python

value_dim = hv.Dimension('Percentage', unit='%')
careers = hv.Sankey((edges, nodes), ['From', 'To'], vdims=value_dim)

careers.opts(
    opts.Sankey(labels='label', label_position='right', width=900, height=300, cmap='Set1',
                edge_color=dim('To').str(), node_color=dim('index').str()))
```
