# Overlap in inputs to oviposition triad
Venn diagrams to show the extent of overlap in input neurons. I will do this for the Hemibrain data as well as for Flywire data to compare the two. I will treat all the subtypes of oviDN as 1 type.

## Hemibrain Venn diagram
There is just 1 rightside oviDNb and 1 oviDNa which have been lumped together here.

```python
# inputs to the oviposition triad
from neuprint import fetch_simple_connections, NeuronCriteria as NC
min_w = 3

oviINr_conns = fetch_simple_connections(None, 'oviIN_R', min_weight=min_w)
oviENr_conns = fetch_simple_connections(None, 'SMP550_R', min_weight=min_w)
oviDNr_conns = fetch_simple_connections(None, NC(instance = 'oviDN.*_R'), min_weight=min_w)

oviDNr_conns
```

```python
oviDNr_conns['bodyId_post'].value_counts()
```

```python
# make lists from the inputs
oviINr_inputs = set(oviINr_conns['bodyId_pre'].tolist())
oviENr_inputs = set(oviENr_conns['bodyId_pre'].tolist())
oviDNr_inputs = set(oviDNr_conns['bodyId_pre'].tolist())
```

```python
# figure showing Venn diagram of overlaps in inputs
from matplotlib_venn import venn3

venn3((oviINr_inputs, oviENr_inputs, oviDNr_inputs), ('oviIN_R', 'oviEN_R','oviDN_R'))
plt.show()
```

## Flywire data
I didn't restrict this analysis to a single hemisphere. Here there were 8 oviDN in the brain.

```python
# import Flywire data
types = pd.read_csv("/Users/ggutierr/Documents/FlyWireData/consolidated_cell_types.csv")
connections_princeton = pd.read_csv("/Users/ggutierr/Documents/FlyWireData/connections_princeton.csv")
```

```python
# group by id pairs to get simple connections, but keep nt_type of the most prevalent synapse group
simple_connections = connections_princeton.sort_values('syn_count', ascending=False)
simple_connections = simple_connections.groupby(['pre_root_id', 'post_root_id']).agg({'syn_count': 'sum', 'nt_type': 'first'}).reset_index()
simple_connections
```

```python
# grab our neurons
oviIN = types[types['primary_type'].str.contains('oviIN', case=False, na=False)|types['additional_type(s)'].str.contains('oviIN', case=False, na=False)]
oviIN
```

```python
oviEN = types[types['primary_type'].str.contains('oviEN|SMP550', case=False, na=False)|types['additional_type(s)'].str.contains('oviEN|SMP550', case=False, na=False)]
oviEN
```

```python
oviDN = types[types['primary_type'].str.contains('oviDN', case=False, na=False)|types['additional_type(s)'].str.contains('oviDN', case=False, na=False)]
oviDN
```

```python
# combine the three ovi neuron dataframes
ovi_neurons = pd.concat([oviIN, oviEN, oviDN]).reset_index(drop=True)
ovi_neurons
```

```python
types[types['primary_type'].str.contains('ovi', case=False, na=False)|types['additional_type(s)'].str.contains('ovi', case=False, na=False)]
```

```python
# inputs to oviIN
oviIN_input_conns = simple_connections[simple_connections['post_root_id'].isin(oviIN['root_id'].tolist())]
oviIN_inputs = set(oviIN_input_conns['pre_root_id'].tolist())
```

```python
# inputs to oviEN
oviEN_input_conns = simple_connections[simple_connections['post_root_id'].isin(oviEN['root_id'].tolist())]
oviEN_inputs = set(oviEN_input_conns['pre_root_id'].tolist())
```

```python
# inputs to oviDN
oviDN_input_conns = simple_connections[simple_connections['post_root_id'].isin(oviDN['root_id'].tolist())]
oviDN_inputs = set(oviDN_input_conns['pre_root_id'].tolist())
```

```python
# Venn diagram of overlaps in inputs
from matplotlib_venn import venn3

plot1 = plt.figure(figsize=(3,3))
venn3((oviIN_inputs, oviEN_inputs, oviDN_inputs), ('oviIN', 'oviEN','oviDN'))
#plt.title('Overlap of inputs to oviposition cell types')
plt.show()
```
